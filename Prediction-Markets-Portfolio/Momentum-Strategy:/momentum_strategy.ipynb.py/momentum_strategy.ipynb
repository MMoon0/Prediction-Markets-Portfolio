```python
# Simulated event-driven momentum example

import pandas as pd
import numpy as np

events = ['Election', 'Policy', 'Sports', 'Climate']
dates = pd.date_range('2025-01-01', periods=12, freq='M')
np.random.seed(42)
predicted_movement = np.random.choice([-1, 1], size=12)  # up or down
actual_movement = predicted_movement.copy()
# introduce some random error
actual_movement[np.random.randint(0,12,3)] *= -1

df = pd.DataFrame({
    'Date': dates,
    'Event_Type': np.random.choice(events, size=12),
    'Predicted': predicted_movement,
    'Actual': actual_movement
})
df['Correct_Prediction'] = df['Predicted'] == df['Actual']
df['Simulated_ROI'] = df['Correct_Prediction'].apply(lambda x: 0.08 if x else -0.04)

print(df)
print("\nTotal simulated ROI:", df['Simulated_ROI'].sum())
```
