```python
# Simulated statistical model example

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
n_simulations = 1000
returns = np.random.normal(0.01, 0.05, n_simulations)  # monthly returns
portfolio_value = np.cumprod(1 + returns/100)

plt.plot(portfolio_value)
plt.title("Simulated Portfolio Growth")
plt.xlabel("Month")
plt.ylabel("Portfolio Value")
plt.show()

print("Expected annual return:", np.mean(returns)*12, "%")
print("Simulated drawdown:", np.min(portfolio_value), "min value")
```
